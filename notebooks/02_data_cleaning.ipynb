```python
from pathlib import Path
import pandas as pd
import numpy as np

PROJECT_ROOT = Path.cwd().parent

RAW_DATA = PROJECT_ROOT / "data" / "raw"
PROCESSED_DATA = PROJECT_ROOT / "data" / "processed"

PROCESSED_DATA.mkdir(exist_ok=True)
```

```python
nav_history = pd.read_csv(RAW_DATA / "nav_history.csv")

transactions = pd.read_csv(RAW_DATA / "investor_transactions.csv")

performance = pd.read_csv(RAW_DATA / "scheme_performance.csv")
```

```python
nav_history.info()
```

```python
# Convert date column to datetime
nav_history["date"] = pd.to_datetime(nav_history["date"])

print(nav_history.dtypes)
```

```python
nav_history = nav_history.sort_values(
    by=["amfi_code", "date"]
).reset_index(drop=True)

nav_history.head()
```

```python
print("Missing NAV values:")
print(nav_history["nav"].isnull().sum())
```

```python
before = len(nav_history)

nav_history = nav_history.drop_duplicates()

after = len(nav_history)

print(f"Duplicates Removed: {before - after}")
```

```python
invalid_nav = nav_history[nav_history["nav"] <= 0]

print("Invalid NAV Records:", len(invalid_nav))
```

```python
nav_history.to_csv(
    PROCESSED_DATA / "nav_history_cleaned.csv",
    index=False
)

print("Saved Successfully!")
```

### Observations - NAV History Dataset

- Successfully converted the date column to datetime format.
- Sorted the dataset by AMFI code and date.
- No missing NAV values were found.
- No duplicate records were detected.
- All NAV values were greater than zero.
- The cleaned dataset was saved to the processed data folder.

```python
performance.columns.tolist()
```

```python
performance.info()
```

```python
numeric_columns = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct",
    "benchmark_3yr_pct",
    "alpha",
    "beta",
    "sharpe_ratio",
    "sortino_ratio",
    "std_dev_ann_pct",
    "max_drawdown_pct",
    "aum_crore",
    "expense_ratio_pct",
    "morningstar_rating"
]

for col in numeric_columns:
    performance[col] = pd.to_numeric(performance[col], errors="coerce")

performance.info()
```

```python
performance[numeric_columns].isnull().sum()
```

```python
invalid_expense = performance[
    (performance["expense_ratio_pct"] < 0.1) |
    (performance["expense_ratio_pct"] > 2.5)
]

print("Invalid Expense Ratio Records:", len(invalid_expense))
```

```python
return_columns = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct"
]

anomalies = performance[
    (performance[return_columns] < -100).any(axis=1) |
    (performance[return_columns] > 200).any(axis=1)
]

print("Anomalous Return Records:", len(anomalies))
```

```python
performance.isnull().sum()
```

```python
before = len(performance)

performance = performance.drop_duplicates()

after = len(performance)

print("Duplicates Removed:", before - after)
```

```python
performance.to_csv(
    PROCESSED_DATA / "scheme_performance_cleaned.csv",
    index=False
)

print("Scheme Performance cleaned successfully!")
```
