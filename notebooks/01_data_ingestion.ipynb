```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent

RAW_DATA = PROJECT_ROOT / "data" / "raw"
```

```python
fund_master = pd.read_csv(RAW_DATA / "fund_master.csv")
nav_history = pd.read_csv(RAW_DATA / "nav_history.csv")

fund_master.head()
```

```python
print("Fund Master Shape:", fund_master.shape)
print("NAV History Shape:", nav_history.shape)

print("\nFund Master Columns")
print(fund_master.columns.tolist())
```

```python
fund_master.isnull().sum()
```

```python
print("Fund Master Duplicate Rows:", fund_master.duplicated().sum())

print("NAV History Duplicate Rows:", nav_history.duplicated().sum())
```

```python
fund_master.columns.tolist()
```

```python
print("Total Schemes:", fund_master.shape[0])
print("Total Columns:", fund_master.shape[1])

print("\nUnique Fund Houses:", fund_master["fund_house"].nunique())
print("Unique Categories:", fund_master["category"].nunique())
print("Unique Sub Categories:", fund_master["sub_category"].nunique())
print("Unique Risk Categories:", fund_master["risk_category"].nunique())
```

```python
print("Minimum AMFI Code:", fund_master["amfi_code"].min())
print("Maximum AMFI Code:", fund_master["amfi_code"].max())
print("Unique Codes:", fund_master["amfi_code"].nunique())
```

```python
master_codes = set(fund_master["amfi_code"])
nav_codes = set(nav_history["amfi_code"])

matching = master_codes.intersection(nav_codes)
missing = master_codes - nav_codes

print("Total Fund Master Codes:", len(master_codes))
print("Total NAV Codes:", len(nav_codes))
print("Matching Codes:", len(matching))
print("Missing Codes:", len(missing))
```

```python
summary = pd.DataFrame({
    "Dataset": ["Fund Master", "NAV History"],
    "Rows": [fund_master.shape[0], nav_history.shape[0]],
    "Columns": [fund_master.shape[1], nav_history.shape[1]],
    "Missing Values": [
        fund_master.isnull().sum().sum(),
        nav_history.isnull().sum().sum()
    ],
    "Duplicate Rows": [
        fund_master.duplicated().sum(),
        nav_history.duplicated().sum()
    ]
})

summary
```

```python
REPORTS = PROJECT_ROOT / "reports"
REPORTS.mkdir(exist_ok=True)

summary.to_csv(REPORTS / "data_quality_summary.csv", index=False)

print("Data Quality Summary saved successfully!")
```
